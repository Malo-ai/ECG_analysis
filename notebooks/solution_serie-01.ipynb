```python
# imports
import numpy as np
import matplotlib.pyplot as plt

# Loading the data file
data = np.loadtxt('../data/ECGu.txt')
ECGs = [data[:, i] * 0.001024 for i in range(3)]  # V to mV conversion

# Creating the time vector
fs = 1000  # Sampling frequency
t = np.arange(len(ECGs[0])) / fs 

plt.style.use('default')
fig, axes = plt.subplots(3, 1, figsize=(12, 10))

# Adding labels to each lead
ylabels = ["Lead I (mV)", "Lead II (mV)", "Lead III (mV)"]

# Plotting each lead
for i, ax in enumerate(axes):
    # ECG grid setup
    ax.grid(True, which='major', color='red', linestyle='-', alpha=0.2)
    ax.grid(True, which='minor', color='lightcoral', linestyle='-', alpha=0.1)
    ax.minorticks_on()
    ax.set_xticks(np.arange(0, 7, 0.2))
    ax.set_xticks(np.arange(0, 7, 0.04), minor=True)  
    ax.set_yticks(np.arange(-1, 1.6, 0.5))   
    ax.set_yticks(np.arange(-1, 1.6, 0.1), minor=True)  

    # Plotting the ECG signal
    ax.plot(t, ECGs[i], 'b-', linewidth=0.8)
    ax.set_ylabel(ylabels[i])
    ax.set_ylim(-1, 1.5)
    ax.set_xlim(0, 6)

    # Adding title to the Plot
    if i == 0:
        ax.set_title("ECGs", pad=20)
    # LLegend for the X-axis only on the last
    if i == len(axes) - 1:
        ax.set_xlabel("Time (s)")

plt.tight_layout()
plt.show()
```
