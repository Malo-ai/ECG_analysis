```python
import numpy as np
import matplotlib.pyplot as plt

# Vérifier le chemin relatif depuis le notebook
ecg = np.loadtxt("../data/ECGu.txt")  
print("Signal chargé avec", len(ecg), "échantillons")
print("10 premières valeurs :", ecg[:10])

# Conversion en millivolts
ecg_mV = ecg * 1.024  # 1024 µV par unité = 1.024 mV
# Axe temps en secondes
fs = 1000  # fréquence d'échantillonnage en Hz
t = np.arange(len(ecg_mV)) / fs  # temps en secondes

# affichage du signal
plt.figure(figsize=(12, 4))
plt.plot(t, ecg_mV)
plt.title("Signal ECG en mV")
plt.xlabel("Temps (s)")
plt.ylabel("Amplitude (mV)")
plt.grid(True)
plt.show()
```
